```python
import requests
from bs4 import BeautifulSoup
import time
from selenium import webdriver
```

```python
#dowload data

file_location = 'productiondata.txt'

print('Beginning file download with requests')

try:
    url = 'http://webapps2.rrc.texas.gov/EWA/productionQueryAction.do'
    r = requests.get(url).text
except Exception as e: 
    print(e)

soup = BeautifulSoup(r,'lxml')
# print(soup.prettify())
print("Complete")
```

```python
for a in soup.find_all('a'):
    print(a)
```

```python
req = requests.get('https://rrcsearch3.neubus.com/esd3-rrc/index.php?_module_=esd&_action_=keysearch&profile=15')

soup = BeautifulSoup(req.text, 'lxml')

print(soup.find_all('input', id_="field_nameTEXT"))

'PHANTOM WOLFCAMP'

response_post = requests.post('https://rrcsearch3.neubus.com/esd3-rrc/index.php?_module_=esd&_action_=keysearch&profile=15',
                           data='PHANTOM WOLFCAMP')
    
print(response_post.text)
```

```python
for n in response_post.find_all("div", id_="overSearchResults"):
    print(n)
```

```python
browser = webdriver.Firefox()

browser.get('https://rrcsearch3.neubus.com/esd3-rrc/index.php?_module_=esd&_action_=keysearch&profile=15')

print("get input")
input_elem = browser.find_element_by_id("field_nameTEXT")

print("send keys")
input_elem.send_keys('PHANTOM WOLFCAMP')

print("click")
browser.find_element_by_id("docSearchButton").click()

time.sleep(6)

docids_tags = browser.find_elements_by_tag_name("tr")

docids = []
for tag in docids_tags:
    docids.append(tag.get_attribute("data-docid"))

print('get all the menus')
action_menus = browser.find_elements_by_class_name("showActionMenu")#.click()
print(len(action_menus))
linkids = []

for menu in action_menus:
    
    print("Open well menu")
    menu.click()
    time.sleep(3)
    
    print("Getting id")
    log_id = menu.find_element_by_xpath("//a[@type='Well Log']").get_attribute("id")
    linkids.append(log_id)
    print(log_id)
    time.sleep(3)
    
    print("close menu")
    menu.find_element_by_xpath("//*[@id='imageMenu']/a[5]").click()
#     browser.find_element_by_class_name("contextMenuClose close").click()
    time.sleep(3)
    
print(docids)
print(linkids)
```
