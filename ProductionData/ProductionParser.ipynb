```python
import os
import csv
import pandas as pd
```

```python
filename = "ProductionDataQuery_Lease_ReportCsv(0).csv"
production_data = {}


for file in os.listdir('.'):
    if file.split(".")[-1] == "csv":
        with open(file) as csv_file:
            csv_reader = csv.reader(csv_file, delimiter=',')
            line_count = 0
            for row in csv_reader:
                if line_count == 5:
                    production_data[file] = [row[0].split("-")[-1]]
                    break
                line_count+=1

        temp_df = pd.read_csv(file,skiprows=7)
        production_data[file].append(temp_df["Oil (BBL)"].sum())
production_data
```

```python
df = pd.DataFrame.from_dict(production_data)
df.head()
```
